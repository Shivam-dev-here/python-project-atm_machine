```python
import time

print("Welcome User, Please insert your card. \n ")
time.sleep(2)
pin = int(input("Enter your security pin number \n"))

print("Please wait Checking Validity")
time.sleep(2)
balance = 5000
if pin == 1234:
    while True:
        print("<-----------------------------------------------Main menu----------------------------------------------->\n")
        print("                  1: Balance          2: Withdraw           3: Deposit           4: Exit \n")
        opt = int(input("choose an option"))

        if opt==1:
            print(f"Your Balance : {balance} INR")
            print("----------------------------------------------------------------------------------------------------------")
        elif opt==2:
            print("Enter amount to cash out")
            amt = int(input("Enter amount"))

            if balance>=amt:
                balance -=amt
                print(f"Your current Balance: {balance}")
                print("----------------------------------------------------------------------------------------------------------")
            else:
                print("Please check your balance before transaction")
                print("----------------------------------------------------------------------------------------------------------")
                break
        elif opt==3:
            amt = int(input("Enter your money"))
            balance +=amt
            print(f"Your current Balance: {balance}")
            print("----------------------------------------------------------------------------------------------------------")

        else:
            break 
else:
    print("Wrong pin, please try again.")
    print("----------------------------------------------------------------------------------------------------------")
    
```
